```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.dpi']= 150

veja = pd.read_csv('veja.csv')
folha = pd.read_csv('folha.csv')
estadao = pd.read_csv('estadao.csv')
uol = pd.read_csv('uol.csv')

keys = ['esporte', 'brasil', 'entretenimento', 'mundo', 'economia', 'politica', 'saude', 'ciencia', 'tecnologia', 'blog', 'placar', 'educacao', 'tveja', 'revista-veja', 'blogs', 'especiais', 'ideias', 'gastronomia']
```

```python
def getColumnsWithThemes(f):
    return list(filter(lambda column: 'Tema' in column, f.columns))

def getColumnsValues(f):
    allThemes = set()
    for column in getColumnsWithThemes(f):
        for theme in f[column]:
            allThemes.add(theme)
    return allThemes


getColumnsValues(uol)
```
